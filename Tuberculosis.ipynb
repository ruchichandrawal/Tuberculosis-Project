```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.lineplot(data=tuberculosis_df, x='Year', y='TB_Incidence_Rate')
plt.title('Global TB Incidence Over Time')
plt.show()
```

```python
sns.boxplot(data=tuberculosis_df, x='Gender', y='TB_Incidence_Rate')
plt.title('TB Incidence by Gender')
plt.show()
```

```python
sns.barplot(data=tuberculosis_df, x='Gender', y='TB_Incidence_Rate')
plt.title('TB Incidence by Gender')
plt.show()
```

```python
tuberculosis_df['Gender'].value_counts().plot.pie(autopct='%1.1f%%')
plt.show()
```
